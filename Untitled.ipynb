```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winemag-data-130k-v2.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['taster_twitter_handle', 'taster_name'], inplace=True)
df.head()
```

```python
df.dropna(subset=['province'], inplace=True)
```

```python
mean_price = df['price'].mean()
df['price'].fillna(mean_price, inplace=True)
```

```python
df.isnull().sum()
```

#### 1. Top Rated Wineries

```python
best_winerys = df.groupby('winery')['points'].mean().round(1)
sorted_grouped_df = best_winerys.sort_values(ascending=False).reset_index()
sorted_grouped_df[:20]
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='winery', y='points', data=sorted_grouped_df[:20], color='b', alpha=0.6)
plt.xticks(rotation=75)
plt.ylim(95, 99)
for index, row in sorted_grouped_df[:20].iterrows():
    plt.text(index, row['points'], round(row['points'], 1), color='black', ha="center")
plt.show()
```

#### 2. Price vs. Quality

```python
top_wineries = sorted_grouped_df.head(20)['winery']
top_wineries_df = df[df['winery'].isin(top_wineries)]
mean_prices = top_wineries_df.groupby('winery')['price'].mean().round(1).reset_index()
top_wineries_with_prices = pd.merge(sorted_grouped_df.head(20), mean_prices, on='winery')
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the Quantity Ordered as barplot on primary y-axis
sns.barplot(x='winery', y='points', data=sorted_grouped_df[:20], ax=ax1, color='b', alpha=0.6)

ax1.set_title('Price vs Quality (based on points)')
plt.ylim(95,99)
for index, row in sorted_grouped_df[:20].iterrows():
    plt.text(index, row['points'], round(row['points'], 1), color='black', ha="center")

# Create a secondary y-axis to overlay the prices
ax2 = ax1.twinx()

# Plot the Price Each as barplot on secondary y-axis
sns.lineplot(x='winery', y='price', data=top_wineries_with_prices, ax=ax2, color='r', alpha=0.6)

# Set label for secondary y-axis

# Rotate x-axis labels for better readability
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=75)

# Display the plot
plt.show()
```

```python
price_quality = df.groupby('winery')['price'].mean().round(1)
sorted_price_qu_df = price_quality.sort_values(ascending=False).reset_index()
sorted_price_qu_df[:10]
```

#### 3. Country and Province Analysis: Determine best provinces in each country

```python
province_mean_points = df.groupby(['country', 'province'])['points'].mean().reset_index()
province_mean_points = province_mean_points.sort_values(['country', 'points'], ascending=[True, False])
best_provinces = province_mean_points.groupby('country').first().reset_index()
best_provinces = best_provinces.sort_values(ascending=False, by='points').reset_index()
best_provinces[:10]
```

#### 4. Determine which wine varieties receive the highest ratings on average

```python
varieties_mean_points = df.groupby('variety')['points'].mean().reset_index()
varieties_mean_points = varieties_mean_points.sort_values(ascending=False, by='points').reset_index()
varieties_mean_points.head(10)
```

```python
top_varieties = varieties_mean_points.head(20)['variety']
top_varieties_df = df[df['variety'].isin(top_varieties)]
mean_prices = top_varieties_df.groupby('variety')['price'].mean().round(1).reset_index()
top_varieties_with_prices = pd.merge(varieties_mean_points.head(20), mean_prices, on='variety')
```

```python
fig, ax3 = plt.subplots(figsize=(10, 6))

# Plot the points as barplot on primary y-axis
sns.barplot(x='variety', y='points', data=varieties_mean_points[:20], ax=ax3, color='b', alpha=0.6)

ax3.set_title('Varieties vs Price (based on points)')
plt.ylim(90, 96)  # Adjusting y-axis limits for better visibility

# Add value labels on top of the bars
for index, row in varieties_mean_points[:20].iterrows():
    plt.text(index, row['points'], round(row['points'], 1), color='black', ha="center")

# Set ticks and labels manually
# ax3.set_xticks(range(len(varieties_mean_points[:20])))
# ax3.set_xticklabels(varieties_mean_points['variety'][:20], rotation=75, ha='right')

# Create a secondary y-axis to overlay the prices
ax4 = ax3.twinx()

# Plot the price as a lineplot on secondary y-axis
sns.lineplot(x='variety', y='price', data=top_varieties_with_prices, ax=ax4, color='r', alpha=0.6, marker='o')

# Set label for secondary y-axis
# ax4.set_ylabel('Price')

# Adjust layout manually to handle margins
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=75)

# Display the plot
plt.show()
```
